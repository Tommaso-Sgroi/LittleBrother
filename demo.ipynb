```python
from ultralytics import YOLO

model = YOLO("yolov8x-oiv7.pt")
```

```python
%matplotlib inline

import matplotlib.pyplot as plt

results = model("img.png", save=True, show_boxes=True, device='cuda')

for result in results:
    plotz = result.plot() # BGR-order numpy array
    plotz = plotz[:, :, ::-1]
    plt.imshow(plotz)
    plt.show()

    result.save_crop('mycrop.jpg', )
```
