```python
from ultralytics import YOLO

model = YOLO("yolo11x.pt")
```

```python
%matplotlib inline

import matplotlib.pyplot as plt

results = model("bus.jpg", save=True, show_boxes=True)

for result in results:
    plotz = result.plot() # BGR-order numpy array
    plotz = plotz[:, :, ::-1]
    plt.imshow(plotz)
    plt.show()
```
